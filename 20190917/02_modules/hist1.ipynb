```python
## 参考 https://stats.biopapyrus.jp/python/boxplot.html

sns.set()
sns.set_style('whitegrid')
sns.set_palette('Set3')

df = pd.DataFrame({
    'variation': x1,
    'same': x2,
    'extreme': x3
})

df_melt = pd.melt(df)
print(df_melt.head())

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
sns.boxplot(x='variable', y='value', data=df_melt, showfliers=False, ax=ax)
sns.stripplot(x='variable', y='value', data=df_melt, jitter=True, color='red', ax=ax)

plt.show()
```
